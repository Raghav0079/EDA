```python
import pandas as pd
import numpy as np

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [1.5, 2.7, 3.0],
    'C': ['apple', 'banana', 'cherry']
})

# Create a dictionary of Series, grouped by dtype
series_dict = {}
for col in df.columns:
    dtype = df[col].dtype
    if dtype not in series_dict:
        series_dict[dtype] = {}
    series_dict[dtype][col] = df[col]

print(series_dict)
```

```python
import json
import csv

# Sample data for the JSON
data = [
    {"name": "Alice", "age": 30, "city": "New York"},
    {"name": "Bob", "age": 25, "city": "Los Angeles"},
    {"name": "Charlie", "age": 35, "city": "Chicago"}
]

# Write the data to a JSON file
with open("data.json", "w") as json_file:
    json.dump(data, json_file, indent=4)  # Use indent for pretty formatting

# Read the JSON file and convert to CSV
with open("data.json", "r") as json_file:
    data = json.load(json_file)

# Extract field names (column headers)
fieldnames = data[0].keys()

with open("data.csv", "w", newline="") as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()  # Write the header row
    writer.writerows(data)  # Write the data rows

print("CSV file 'data.csv' created successfully!")
```

```python
def evenoddsum(lst):
    even_sum = 0
    odd_sum = 0
    for i in lst:
        if i % 2 == 0:  # Corrected the if condition
            even_sum = even_sum + i
        else:
            odd_sum = odd_sum + i

    return even_sum, odd_sum

lst = [1, 2, 3, 4, 5, 6]  # Example list
even_sum, odd_sum = evenoddsum(lst)
print("Even sum:", even_sum)
print("Odd sum:", odd_sum)
```

```python
def lst_sq(lst):
    """
    This function takes a list of numbers and returns a new list with the squares of those numbers.
    """
    result = []
    for num in lst:
        result.append(num ** 2)
    return result

lst = [1, 3, 45, 6, 7, 2]
squared_lst = lst_sq(lst)
print(squared_lst)  # Output: [1, 9, 2025, 36, 49, 4]
```
